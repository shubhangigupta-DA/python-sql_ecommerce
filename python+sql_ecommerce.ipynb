# 3. Find the total sales per category.

```python
import mysql.connector

# Establish MySQL connection
conn = mysql.connector.connect(
    host='localhost',
    user='root',  # Corrected parameter
    password='SGUPTA@1234!',
    database='ecommerce'
)

cur = conn.cursor()

# Corrected SQL Query
query = """
SELECT products.product_category AS category, SUM(payments.payment_value) AS sales
FROM products
JOIN order_items ON products.product_id = order_items.product_id  -- Corrected Join Condition
JOIN payments ON payments.order_id = order_items.order_id
GROUP BY category
"""

cur.execute(query)

data = cur.fetchall()

# Loop through results
for row in data:
    print(f"Category: {row[0]}, Sales: {row[1]}")

cur.close()
conn.close()
```
